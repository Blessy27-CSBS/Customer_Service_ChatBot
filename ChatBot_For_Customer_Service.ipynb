Importing Required Libraries:

```python
import random
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

```

Define Intents 

```python
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "responses": [
            "Hello! How can I assist you today?",
            "Hi there! What can I help you with?"
        ]
    },
    
    "order_status": {
        "patterns": ["order", "status", "track"],
        "responses": [
            "Please provide your order ID to track your order.",
            "You can track your order using your order ID."
        ]
    },
    
    "refund": {
        "patterns": ["refund", "money", "cancel"],
        "responses": [
            "Refunds are processed within 5–7 business days.",
            "Your refund request has been noted."
        ]
    },
    
    "contact": {
        "patterns": ["contact", "support", "help"],
        "responses": [
            "You can contact support at support@email.com"
        ]
    },
    
    "hours": {
        "patterns": ["hours", "open", "time"],
        "responses": [
            "We are open 9 AM to 6 PM, Monday to Friday."
        ]
    },
    
    "goodbye": {
        "patterns": ["bye", "goodbye"],
        "responses": [
            "Goodbye! Have a great day!",
            "See you again!"
        ]
    }
}
```

Response Function

```python
def get_response(message):
    tokens = word_tokenize(message.lower())
    
    for intent in intents.values():
        for pattern in intent["patterns"]:
            if pattern in tokens:
                return random.choice(intent["responses"])
    
    return "Sorry, I didn't understand. Can you rephrase?"
```

Chat Loop

```python
print("Customer Service Bot Ready (type 'quit' to exit)\n")

while True:
    msg = input("You: ")
    
    if msg.lower() == "quit":
        print("Bot: Goodbye!")
        break
        
    print("Bot:", get_response(msg))
```

```python
import gradio as gr

def bot_logic(message, history):

    msg = message.lower()

    if "hi" in msg or "hello" in msg:
        reply = "Hello! How can I help you today?"
    elif "refund" in msg:
        reply = "Refunds take 5–7 business days."
    elif "order" in msg:
        reply = "Please share your order ID to track it."
    elif "contact" in msg:
        reply = "Contact support@email.com"
    else:
        reply = "Sorry, I didn't understand."

    # Add messages in new format
    history.append({"role": "user", "content": message})
    history.append({"role": "assistant", "content": reply})

    return "", history


with gr.Blocks() as demo:
    gr.Markdown("<h2 style='text-align:center'>Customer Service Chatbot</h2>")

    chatbot = gr.Chatbot(type="messages", height=400)

    msg = gr.Textbox(placeholder="Type your message...")

    clear = gr.Button("Clear Chat")

    msg.submit(bot_logic, [msg, chatbot], [msg, chatbot])
    clear.click(lambda: [], None, chatbot)

demo.launch()
```
